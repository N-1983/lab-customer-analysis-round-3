```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.__version__
```

```python
file = pd.read_csv('marketing_customer_analysis.csv')
file
```

```python
file.describe()
```

```python
file['Response'].value_counts().plot(kind='bar')
#Most of the Customers are not responded.
```

```python
Active = file[file['Response']== 'Yes']
Active
#extract customers who are responded.
```

```python
sns.countplot(x='Sales Channel',hue='Response',data =Active)
#Most of the customer's responded by Agents.
```

```python
sns.boxplot(x=Active['Total Claim Amount'])
# The total claim of the active responded customers are paying between $300 and $600
```

```python
sns.displot(Active['Income'],bins=20)
# Those who have less income they are very responded well than highly income customer's.
```
